```python
import pickle
import numpy as np
from sklearn.preprocessing import RobustScaler,MinMaxScaler
X_ppg,X_qual,X_acc,y_rr, y_respiration,y_inspiration,y_expiration,groups,X_respiration = pickle.load(open('../../affect-recognition/data/tabular_data.p','rb'))
```

```python
from scipy import signal
X_ppg = X_ppg.reshape(-1,512,1)
X_ppg = np.array([RobustScaler().fit_transform(a) for a in X_ppg])
X_respiration = np.array([RobustScaler().fit_transform(signal.resample(a.reshape(-1),512).reshape(-1,1)) for a in X_respiration])
# y_rr = 60000/(1000*y_rr)
print(X_ppg.shape,X_respiration.shape)
X_respiration = X_ppg
```

```python
# train a generative adversarial network on a one-dimensional function
from numpy import hstack
from numpy import zeros
from numpy import ones
from numpy.random import rand
from numpy.random import randn
from keras.models import Sequential,Input
from keras.layers import Input, LSTM, RepeatVector,Bidirectional,Multiply,multiply,Permute
from keras.layers import TimeDistributed,Dense,Flatten,Reshape,Lambda,Activation,GRU
from keras.models import Model
from keras.callbacks import ModelCheckpoint
from keras.callbacks import EarlyStopping
from keras.layers import Dense
from matplotlib import pyplot
%matplotlib inline

# define the standalone discriminator model
def define_discriminator(n_inputs=512,n_dim=1):
    inputs = Input(shape=(n_inputs, n_dim))
    reshaped_input = Reshape((n_inputs,))(inputs)
    layer1 = Dense(1, activation='selu', kernel_initializer='normal', input_dim=n_inputs)(reshaped_input)
#     layer1 = Dense(500, activation='selu', kernel_initializer='normal', input_dim=n_inputs)(layer1)
    layer1 = Dense(25, activation='selu', kernel_initializer='normal', input_dim=n_inputs)(layer1)
    output = Dense(1, activation='sigmoid')(layer1)
    # compile model
    model = Model(inputs=[inputs], outputs=[output])
    model.compile(loss='mae', optimizer='adam', metrics=['accuracy'])
    return model

# define the standalone generator model
def define_generator(n_inputs = 512,n_outputs=512,n_dim=1,first_dim=5):
    inputs = Input(shape=(n_inputs, n_dim))
#     reshaped_input = Reshape((n_inputs,))(inputs)
    layer1 = Bidirectional(GRU(first_dim,return_sequences=True,activation='linear',go_backwards=True))(inputs)
    layer1 = TimeDistributed(Dense(1, activation='linear'))(layer1)
#     layer1 = Flatten()(layer1)
#     layer1 = Dense(n_inputs,activation='tanh')(layer1)
#     layer1 = Reshape((n_inputs,n_dim))(layer1)
    layer1 = GRU(1, activation='linear',return_sequences=True)(layer1)
#     layer1 = Dense(2000, activation='sigmoid', kernel_initializer='he_uniform', input_dim=n_inputs)(layer1)
#     output = Dense(n_outputs, activation='sigmoid')(layer1)
#     output = Reshape((n_outputs,n_dim))(output)
    model = Model(inputs=[inputs], outputs=[layer1])
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# define the combined generator and discriminator model, for updating the generator
def define_gan(generator, discriminator,n_inputs = 512,n_outputs=1,n_dim=1):
    # make weights in the discriminator not trainable
    inputs = Input(shape=(n_inputs, n_dim))
    gen_output = generator(inputs)
    dis_output = discriminator(gen_output)
    discriminator.trainable = False
    model = Model(inputs=[inputs], outputs=[dis_output,gen_output])
#     losses = 
    print(model.layers[1].name,model.layers[2].name)
    losses ={model.layers[1].name:'mse',
          model.layers[2].name:'binary_crossentropy',    
        }
    # connect them
    # compile model
    lossWeights={model.layers[1].name:10,
          model.layers[2].name:10,   
        }
    model.compile(loss=losses, optimizer='adam',loss_weights=lossWeights)
    return model

# generate n real samples with class labels
def generate_real_samples(n_batch):
    # generate inputs in [-0.5, 0.5]
    index = np.arange(X_respiration.shape[0]).reshape(-1)
    index_picked = np.random.choice(index,n_batch)
    X = X_respiration[index_picked]
    y = ones((n_batch, 1))
    return X, y

# generate points in latent space as input for the generator
def generate_latent_points(latent_dim=512, n_batch=1000):
    # generate points in the latent space
    index = np.arange(X_ppg.shape[0]).reshape(-1)
    index_picked = np.random.choice(index,n_batch)
    X = X_ppg[index_picked]
    return X,X_respiration[index_picked]

# use the generator to generate n fake examples, with class labels
def generate_fake_samples(generator, latent_dim=512, n=1000):
    # generate points in latent space
    x_input,x_respiration = generate_latent_points(latent_dim, n)
    # predict outputs
    X = generator.predict(x_input)
    print('generator performance')
    pyplot.plot(x_input[0],'r')
    pyplot.plot(x_respiration[0],'g')
    pyplot.plot(RobustScaler().fit_transform(X[0]),'b')
#     pyplot.scatter(x_fake[:, 0], x_fake[:, 1], color='blue')
    pyplot.show()
    # create class labels
    y = zeros((n, 1))
    return X, y

# evaluate the discriminator and plot real and fake points
def summarize_performance(epoch, generator, discriminator, latent_dim, n=1000):
    # prepare real samples
    x_real, y_real = generate_real_samples(n)
    # evaluate discriminator on real examples
    _, acc_real = discriminator.evaluate(x_real, y_real, verbose=0)
    # prepare fake examples
    x_fake, y_fake = generate_fake_samples(generator, latent_dim, n)
    # evaluate discriminator on fake examples
    _, acc_fake = discriminator.evaluate(x_fake, y_fake, verbose=0)
    # summarize discriminator performance
    print(epoch, acc_real, acc_fake)
    # scatter plot real and fake data pointspyplot.figure(figsize=(10,5))
    pyplot.plot(x_real[0],'r')
    pyplot.plot(x_fake[0],'b')
#     pyplot.scatter(x_fake[:, 0], x_fake[:, 1], color='blue')
    pyplot.show()

# train the generator and discriminator
def train(g_model, d_model, gan_model, latent_dim=5590, n_epochs=1000, n_batch=400, n_eval=1):
    # determine half the size of one batch, for updating the discriminator
    half_batch = int(n_batch / 2)
    # manually enumerate epochs
    for i in range(n_epochs):
        # prepare real samples
        x_real, y_real = generate_real_samples(half_batch)
        # prepare fake examples
        x_fake, y_fake = generate_fake_samples(g_model, latent_dim, half_batch)
        print(x_real.shape,y_real.shape,x_fake.shape,y_fake.shape)
        # update discriminator
        print('training')
        d_model.train_on_batch(np.concatenate([x_real,x_fake]), np.concatenate([y_real,y_fake]))
#         d_model.train_on_batch(x_fake, y_fake)
        # prepare points in latent space as input for the generator
        x_gan,x_resp_gan = generate_latent_points(latent_dim, n_batch)
        # create inverted labels for the fake samples
        y_gan = ones((n_batch, 1))
        print(x_gan.shape)
        # update the generator via the discriminator's error
        gan_model.train_on_batch(x_gan, [y_gan,x_resp_gan])
#         evaluate the model every n_eval epochs
#         if (i+1) % n_eval == 0:
#             summarize_performance(i, g_model, d_model, latent_dim)

# size of the latent space
latent_dim = 512
# create the discriminator
discriminator = define_discriminator()
discriminator.summary()
# create the generator
generator = define_generator()
generator.summary()
# # create the gan
gan_model = define_gan(generator, discriminator)
gan_model.summary()
# # train model
train(generator, discriminator, gan_model, latent_dim)
```
